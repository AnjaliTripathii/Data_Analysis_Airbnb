```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\compressed_data.csv\compressed_data.csv")
```

Checking null values


```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month' : 0 , 'last review': df['last review'].min()}, inplace = True)
```

```python
df.dropna( subset= ['NAME', 'host name'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ['license', 'house_rules'], errors = 'ignore')
```

```python
df.columns
```

```python
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

```python
df.drop_duplicates( inplace = True )
```

VISUALISATION


What is the distribution of listing prices?

```python
plt.figure(figsize = (10,5))
sns.histplot(df['price'], bins=50, kde=True, color='green')
plt.title('Distribution of listing price')
plt.xlabel('Price in $')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows a fairly even distribution of listing prices across different price ranges indicating no particualr concentration of listings in any specific price range.

How are different room types distributed?

```python
plt.figure(figsize=(8,5))
sns.countplot( x='room type', data=df, color='lightblue')
plt.title('Room type distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

How are listings distributed across different neighbourhood?

```python
plt.figure(figsize=(12,8))
sns.countplot( y='neighbourhood group', data=df, color='hotpink', order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by neighbourhood group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

What is the relation between price and room tyoe?

```python
plt.figure(figsize=(10,6))
sns.boxplot( x='room type', y='price', hue='room type', data=df, palette='Set1')
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price($)')
plt.legend(title='Room type')
plt.show()
```

How has the number of reviews changed over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line', color='brown')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.ylabel('Number of reviews')
plt.legend(title='Room type')
plt.show()
```
